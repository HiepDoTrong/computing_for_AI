### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic 

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv'
titanic = pd.read_csv(url)
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic = titanic.set_index('PassengerId')
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
import matplotlib.pyplot as plt

males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

proportions = [males, females]

plt.figure(figsize=(8, 6)) 
plt.pie(proportions, labels=['Males', 'Females'], colors=['blue', 'green'], explode=(0.15, 0), startangle=90, autopct='%1.1f%%', shadow=False)

plt.axis('equal') 
plt.title("Sex Proportion")
plt.legend(['Males', 'Females'], loc='upper right')
plt.tight_layout()  
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
import seaborn as sns

lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex')
lm.set(title = 'Fare / Age')
axes = lm.axes
axes[0,0].set_ylim(0,)
axes[0,0].set_xlim(0,85)
```

### Step 7. How many people survived?

### Step 8. Create a histogram with the Fare payed

```python
import numpy as np

df = titanic.Fare.sort_values(ascending = False)
binsVal = np.arange(0,600,10)
plt.hist(df, bins = binsVal)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Payed')
plt.show()
```
